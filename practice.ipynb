```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data for heatmap
heatmap_data = np.random.rand(10, 10)

# Generate random data for scatter plot
x = np.random.rand(50)
y = np.random.rand(50)

# Create a figure with the specified size
fig, ax = plt.subplots(figsize=(7, 6))

# Create the heatmap at the center of the figshape
heatmap = ax.imshow(heatmap_data, cmap='viridis', interpolation='nearest', extent=(2, 8, 2, 8))

# Add a colorbar to the heatmap
cbar = plt.colorbar(heatmap, ax=ax)

# Overlay scatter plot on top of the heatmap
ax.scatter(x * 10, y * 10, c='red')

# Set the title and labels
ax.set_title('Random Data Heatmap with Scatter Overlay')
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')

# Show the plot
plt.show()
```
