## Copying same files of two different folders

```python

# Source folders
folder1 = r"""C:\Users\asad\Desktop\Minahil\Images"""
folder2 = r"""C:\Users\asad\Desktop\Minahil\Labels"""

# Destination folder
destination_folder = r"""C:\Users\asad\Desktop\Minahil_new"""

# Get list of files in each folder
files1 = os.listdir(folder1)
files2 = os.listdir(folder2)

# Match files from folder1 with files in folder2 based on filename
for file1 in files1:
    for file2 in files2:
        # Extract filenames without extensions
        filename1, _ = os.path.splitext(file1)
        filename2, _ = os.path.splitext(file2)

        if filename1 == filename2:
            # Source paths for files
            source_file1 = os.path.join(folder1, file1)
            source_file2 = os.path.join(folder2, file2)
            
            # Destination paths for files
            destination_file1 = os.path.join(destination_folder, file1)
            destination_file2 = os.path.join(destination_folder, file2)

            # Copy or move files to the destination folder
            shutil.copy(source_file1, destination_file1)  # Change copy to move if you want to move files
            shutil.copy(source_file2, destination_file2)  # Change copy to move if you want to move files
        
```

## Segment to Box


```python
def seg_to_bbox(seg_info):
    # Example input: 2 0.207031 0.558594 0.208984 0.527344 0.210938 0.488281 0.214844 0.445312 0.21875 0.412109 0.222656 0.382812
    class_id, *points = seg_info.split()
    points = [float(p) for p in points]
    x_min, y_min, x_max, y_max = min(points[0::2]), min(points[1::2]), max(points[0::2]), max(points[1::2])
    bbox_info = f"{class_id} {x_min} {y_min} {x_max} {y_max}"
    return bbox_info
```
